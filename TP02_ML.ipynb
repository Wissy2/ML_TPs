```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder

# Sample DataFrame (replace this with your actual BankD DataFrame)
# BankD = pd.read_csv("your_dataset.csv")  # Make sure to load your actual data

# Define categorical columns
ordinal_cols = ['Education_Level', 'Income_Category']  # Ordered categories
nominal_cols = ['Gender', 'Marital_Status', 'Card_Category']  # Unordered categories

# Define the correct order for ordinal categories
education_levels = ['Uneducated', 'High School', 'College', 'Graduate', 'Post-Graduate', 'Doctorate', 'Unknown']
income_levels = ['Less than $40K', '$40K - $60K', '$60K - $80K', '$80K - $120K', '$120K +', 'Unknown']

# Apply ordinal encoding for Education_Level and Income_Category
ordinal_encoder = OrdinalEncoder(categories=[education_levels, income_levels])
BankD[ordinal_cols] = ordinal_encoder.fit_transform(BankD[ordinal_cols])

# Apply Label Encoding for 'Gender' (binary categories)
label_encoder = LabelEncoder()
BankD['Gender'] = label_encoder.fit_transform(BankD['Gender'])

# Apply One-Hot Encoding for 'Marital_Status' and 'Card_Category'
BankD = pd.get_dummies(BankD, columns=['Marital_Status', 'Card_Category'], drop_first=True)

# Check transformed dataset
print(BankD.head())



```

```python
# Convert Attrition_Flag to binary (0 = Existing Customer, 1 = Attrited Customer)
BankD['Attrition_Flag'] = BankD['Attrition_Flag'].map({'Existing Customer': 0, 'Attrited Customer': 1})
BankD
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Define continuous numerical columns
numeric_cols = ['Customer_Age', 'Income_Category', 'Total_Trans_Amt', 'Total_Trans_Ct', 'Avg_Utilization_Ratio']

# Plot histograms for the continuous variables
plt.figure(figsize=(12, 10))

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[col], kde=True)  # Kernel Density Estimate adds smooth line
    plt.title(f'Histogram of {col}')

plt.tight_layout()
plt.show()

```

```python
# Cap outliers to the lower and upper bounds
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap outliers
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

print("Outliers have been capped.")
```

```python
import numpy as np  # Import the NumPy library

# Replace outliers with the median
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 =df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Replace outliers with median
    median_value = BankD[col].median()
    df[col] = np.where((BankD[col] < lower_bound) | (df[col] > upper_bound), median_value, df[col])

print("Outliers have been replaced with the median.")
```

```python
import numpy as np

# Apply log transformation for skewed columns
for col in numeric_cols:
    df[col] = np.log1p(df[col])  # log1p handles 0 and negative values

print("Log transformation applied.")
```

```python
# Plot box plots to detect outliers
plt.figure(figsize=(12, 10))

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')

plt.tight_layout()
plt.show()
```
